```python
from bs4 import BeautifulSoup
import requests
url  = BeautifulSoup('https://www.espn.com/soccer/standings/_/league/eng.1', 'html.parser')
soup = requests.get(url)
soup = BeautifulSoup(soup.text, "lxml")
soup
```

```python
table = soup.css.select(".standings__table InnerLayout__child--dividers")
tags = table.find_all('tr')
tags
```
