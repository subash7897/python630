```python
from bs4 import BeautifulSoup
import requests
url  = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup = requests.get(url)
soup = BeautifulSoup(soup.text, "lxml")
soup
```

```python
table=soup.table
table
```

```python
tags= table.find_all('tr')
```

```python
data = []
for i in tags[1:]:
    x= i.text.split('\n')[1:23]
    if x[0] != '':
        data.append(x)

print(data)
```

```python
heading = tags[0].text.split('\n')[1:23]
heading
```

```python
import pandas as pd
df= pd.DataFrame(data,columns=heading)
df.to_csv('Coviddata.csv')
df
```
